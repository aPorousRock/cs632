# Set index for train and test

```python
# Split data into train and test
train_size = int(len(body_file) * .5)
print ("Train size: %d" % train_size)
print ("Test size: %d" % (len(body_file) - train_size))
```

# Prepare train and test data in 50:50 ratio

```python

train_posts = body_file['post'][:train_size]
train_tags = label_file['label'][:train_size]

test_posts = body_file['post'][train_size:]
test_tags = label_file['label'][train_size:]
```

# Set bag of words array length

```python
max_words = 100
tokenize = text.Tokenizer(num_words=max_words, char_level=False)
```

# train posts

```python
tokenize.fit_on_texts(train_posts) # only fit on train
```

```python
text = "tr"
# tokenize the text, and retrieve the index of the first 10 tokens
tokenize.texts_to_matrix([text])
```

```python
x_train = tokenize.texts_to_matrix(train_posts)
x_test = tokenize.texts_to_matrix(test_posts)
```

```python
#print(tokenize.word_index)
x_train[0]
```

```python
encoder = LabelEncoder()
encoder.fit(train_tags)
y_train = encoder.transform(train_tags)
y_test = encoder.transform(test_tags)
```

```python
num_classes = np.max(y_train) + 1
y_train = utils.to_categorical(y_train, num_classes)
y_test = utils.to_categorical(y_test, num_classes)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
num_classes
```

```python
batch_size = 5
epochs = 10
```

```python
model = Sequential()
model.add(Dense(512, input_shape=(max_words,)))
model.add(Activation('relu'))
model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

```

```python
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split=0.1)
```

```python
score = model.evaluate(x_test, y_test,
                       batch_size=batch_size, verbose=1)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
from scipy.spatial import distance
def eucledean(a,b):
        return distance.euclidean(a,b)
```

```python
class myCustomClassifier():
    
    
    def fit(self,x_train,y_train):
        self.x_train = x_train
        self.y_train = y_train
        
    def closest(self,row):
        closest_distance = eucledean(row,self.x_train[0])
        closest_index = 0
        for i in range(1,len(x_train)):
            
            dist = eucledean(row,self.x_train[i])
            if dist < closest_distance:
                closest_distance = dist
                closest_index = i
                
        return self.y_train[closest_index]
        
    def predict(self,x_test):
        predictions = []
        for row in x_test:
            #print(row)
            label = self.closest(row)
            predictions.append(label)
        return predictions
```

```python
knn = myCustomClassifier()
knn.fit(x_train, y_train) 
```

```python
Final_predictions = knn.predict(x_test)
print(len(Final_predictions))
print(Final_predictions)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,Final_predictions))
```
